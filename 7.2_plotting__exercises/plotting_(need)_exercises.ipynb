```python
#task 1
import matplotlib.pyplot as plt

# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='maroon', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc = 'lower right')
plt.show()
```

```python
#task 2 - 1
import numpy as np
x = np.arange(0,51)
y = (np.e**-(x/10))*np.sin(x)


plt.plot(x,y, 
         label='Y value')
plt.xlabel('x')
plt.ylabel('y')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc = 'lower right')
```

```python
#task 2 - 2
import numpy as np
x = np.arange(0,51)
x = np.linspace(0, 51, 200)
y = (np.e**(-x/10))*np.sin(x)
plt.ylim(-1,1)

plt.plot(x,y)
plt.xlabel('x',fontsize=15,color='darkorange')
plt.ylabel('y',fontsize=15,color='darkorange',rotation = 360)
plt.grid(alpha=0.25)

plt.title(r'$y = e^{-x/10}\sin(x)$',fontsize=20)
# a pair of '$" is to enclose the whole function 
#LaTex need \ before special math functions like sin(x)

plt.grid(alpha=.25)
```
