```python
#task 4
import glob
import os

convert_filename = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
}

def rename_my_file(old_file_name):
    month = old_file_name.split("-")[0]
    
    month_num = convert_filename[month]

    new_name = old_file_name.replace(month, month_num)
    new_name.replace("_","-")
    return new_name

path = os.path.join('os-collaboration-exercise-data', '*.txt')
file_A = glob.glob(path)
new_list_of_files = []

for files in file_A:
    file_to_be_renamed = files.split(os.path.sep)[1]
    #files in file A will be os-collaboration-exercise-data/xxx, but we only want xxx part
    #separators are different in different systems, so use os.path.sep to refer to separator
    new_list_of_files.append(file_to_be_renamed)


for index, files in enumerate(new_list_of_files):
    try:
        new_name = rename_my_file(files)
        os.rename(file_A[index], os.path.join('os-collaboration-exercise-data', new_name))

    except:
        print('everything is renamed')
    
```

```python
#task 6
import shutil

Cairo_city = glob.glob('os-collaboration-exercise-data/*cairo*')
Cairo_path = os.path.join('cities', 'Cairo')

for files in Cairo_city:
    shutil.copy(files, Cairo_path)
```

```python
#task 7
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
    'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for city_moved in cities:
    city_list = glob.glob(f'os-collaboration-exercise-data/*{city_moved.lower()}*')
    city_path = os.path.join('cities', city_moved)
    
    for files in city_list:
        try:
            shutil.move(files, city_path)
        except:
            print('everything is moved')
    
```
